```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
```

```python
x=np.linspace(0,5,11)
```

```python
y=x**2
```

```python
x
```

```python
y
```

```python
#functional
plt.plot(x,y)
plt.xlabel('X label')
plt.ylabel('Y Label')
plt.title('Title')
```

```python
 plt.subplot(1,2,2)
plt.plot(x,y,'r')
plt.subplot(1,2,1)
plt.plot(y,x,'b')
```

```python
#Object Oriented
fig=plt.figure()

axes=fig.add_axes([0.1,0.3,0.8,0.8])
axes.plot(x,y)
axes.set_xlabel('X label')
axes.set_ylabel('Y label')
axes.set_title('Set Title')
```

```python
fig=plt.figure()
axes1=fig.add_axes([0.1,0.1,0.8,0.8])
axes2=fig.add_axes([0.2,0.5,0.4,0.3])
axes1.plot(x,y)
axes1.set_title('Larger Plot')
axes2.plot(y,x)
axes2.set_title('Smaller Plot')
```

```python
fig,axes=plt.subplots(1,2)
axes[0].plot(x,y)
axes[0].set_title("First")
axes[1].plot(y,x)
axes[1].set_title("Second")

plt.tight_layout()
```

# Figure Size and DPI

```python
# fig=plt.figure(figsize=(8,2))
# ax=fig.add_axes([0.1,0.1,0.8,0.8])
# ax.plot(x,y)

fig,axes=plt.subplots(2,1,figsize=(8,2))
axes[0].plot(x,y)
axes[1].plot(y,x)
plt.tight_layout()
```

```python
fig.savefig('my_picture.png',dpi=200)
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,x**2,label='X squared')
ax.plot(x,x**3,label='X cubed')
ax.legend()
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y,color='purple',lw=1,ls='--') 
ax.set_xlim([0,1])
ax.set_ylim([0,2])
# ax.hist(x,y)
```
